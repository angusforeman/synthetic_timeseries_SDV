```python
import pandas

master_input_data = pandas.read_csv('./master_concatenated_stock_input_data.csv')
```

```python
import matplotlib.pyplot as plt

# plot the synth data grouped by stock symbol
grouped_data = master_input_data.groupby('Symbol')

# plot the data for each stock symbol
for symbol, data in grouped_data:
    data.plot(x='Combineddatetime', y='Open', label="Source data for " + symbol)

# add a legend to the plot

plt.legend()
plt.show()
```
