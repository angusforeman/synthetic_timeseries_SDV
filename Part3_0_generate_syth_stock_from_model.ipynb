```python
from sdv.sequential import PARSynthesizer
synthesizer = PARSynthesizer.load('IntraDayStockSynthesiser.pkl') # Load the model we trained  in step 2
```

```python
synthetic_data = synthesizer.sample(num_sequences=12, sequence_length=300) #generate 12 sequences (of Stock Symbols) of 96 rows each (8 hours worth of 5 minute data samples)
synthetic_data.head()
```

```python
#sort the dataset so we can better see the sequences (which in my experience are not always correct for time series)
synthetic_data.sort_values(['Symbol', 'Combineddatetime'])
```

```python
synthetic_data.to_csv('synthetic_Intraday_PARstock.csv', index=False) #save the synthetic data to a csv file
synthetic_data['Symbol'].unique() #list the synthetic stock symbols
```
